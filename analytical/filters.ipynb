```python
# load an image located at "orig.png"
import cv2 as cv
import numpy as np
import scipy
from IPython.display import Image
img = cv.imread("orig.png")
#display image
Image("orig.png")
```

```python
box_blur_kernel = np.array([[1, 1, 1],
                            [1, 1, 1],
                            [1, 1, 1]]) / 9.0

#convolve image with box blur kernel

box_blur_img = cv.filter2D(img, -1, box_blur_kernel)

# write out image to "box_blur_test.png"
cv.imwrite("box_blur_test.png", box_blur_img)
Image("box_blur_test.png")
```

```python
# take a median filter of the img
median_filtered = cv.medianBlur(img, 3)
cv.imwrite("median_filtered.png", median_filtered)
Image("median_filtered.png")
```
